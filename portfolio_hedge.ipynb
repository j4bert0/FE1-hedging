# Delta hedging (portfolio of options)

When delta hedging a portfolio of options two cases are considered:
* When the underlying is the same for all of the options in the portfolio

$$\Delta = \frac{\partial P}{\partial S}= \frac{\partial}{\partial S}\bigg(w_1C_t^1+w_2C_t^2+...+w_KC_t^K \bigg)=\sum_{i=1}^Kw_i\Delta_i$$

where $P=$ portfolio of options, $w_i=$ weight of option i, $C_t^i=$ price of i at time t and $\Delta_i=$ delta of options i

* When the underlying is not the same
